## Song Breakdown per League

```python
import pandas as pd
```

```python
players = pd.read_csv("cleaned_players.csv", index_col = "player_no")
songs = pd.read_csv("Major League WalkUp Song Playlist_20230721 - songs.csv")
teams = pd.read_csv("Major League WalkUp Song Playlist_20230721 - teams.csv")
player_song = players.merge(songs, how = "inner", on = "song_code")
player_song.head(3)
```

```python
team_player_song = player_song.merge(teams, how = "inner", on = "team_code")
team_player_song.head(3)
```

```python
grouping_by_league = team_player_song.groupby(["song_name", "artist_name", "league"])
```

```python
grouping_by_league.size()
```

```python
sl = pd.DataFrame(grouping_by_league.size().sort_values(ascending = False))
```

```python
sl.head(5)
```

## Song Breakdown per Division
